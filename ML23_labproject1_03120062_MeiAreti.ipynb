3.α.  Θα χρησιμοποιήσετε όλα τα χαρακτηριστικά του συνόλου δεδομένων για την εκπαίδευση των ταξινομητών ή θα επιλέξετε κάποια από αυτά;  Θα κάνετε κάποια συνένωση κάποιων χαρακτηριστικών για δημιουργία νέων χαρακτηριστικών για το μοντέλο σας;


Όσον αφορά στα χαρακτηριστικά που θέλουμε να επιλέξουμε, θα πρέπει να πληρούνται κάποιες προυποθέσεις, όπως:
Σημασιολογική Σημασία: Χαρακτηριστικά που έχουν σημασιολογική σημασία για το πρόβλημα.

Συσχέτιση με τον Στόχο: Επιλέγουμε χαρακτηριστικά που έχουν σχέση με τον στόχο της ταξινόμησης και αναλύουμε τα χαρακτηριστικά ως προς το πώς σχετίζονται με την ετικέτα που προσπαθούμε να προβλέψουμε.

Εξαλειφή Συνεκτικότητας: Καλό είναι να αποφευχθεί η επιλογή χαρακτηριστικών που είναι γραμμικά συνεκτικά (collinear), δηλαδή που σχετίζονται σε μεγάλο βαθμό μεταξύ τους. Αυτό μπορεί να οδηγήσει σε προβλήματα όταν χρησιμοποιούνται αλγόριθμοι που βασίζονται στην αντίστροφη του πίνακα.

Διαχείριση Δεδομένων Εξόδου: Θα χρειαστεί να εξετάσουμε τον τύπο των δεδομένων. Σε περίπτωση κατηγορικών χαρακτηριστικών, θα πρέπει να μετατραπούν σε αριθμητική μορφή.

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
import pandas as pd
import numpy as np
```

```python
# Set a threshold for high correlation
threshold = 0.9

numeric_features = df.select_dtypes(include=np.number)

# Compute the correlation matrices for numeric features
numeric_corr_matrix = numeric_features.corr()

# Find pairs with correlation above the threshold for numeric features, one feature of each pair must be excluded
high_corr_numeric_pairs = np.where(np.abs(numeric_corr_matrix) > threshold)

# Create sets to store unique pairs of numeric features
unique_numeric_pairs = set()

# Iterate through the numeric pairs and add them to the set
for i, j in zip(high_corr_numeric_pairs[0], high_corr_numeric_pairs[1]):
    # Exclude diagonal elements
    if i < j:
        # Get the corresponding numeric feature names
        feature_i = numeric_features.columns[i]
        feature_j = numeric_features.columns[j]

        # Add a tuple of numeric feature names to the set
        unique_numeric_pairs.add(tuple(sorted([feature_i, feature_j])))


print("Pairs of features with high correlativity", unique_numeric_pairs)
# Create a copy of the original DataFrame
df_copy = df.copy()

# Function to drop one feature from each highly correlated pair
def drop_highly_correlated_features(dataframe, corr_pairs):
    for pair in corr_pairs:
        feature_i, feature_j = pair
        # Choose the feature with higher absolute correlation with the target variable
        correlation_i = np.abs(dataframe[feature_i].corr(dataframe['RainTomorrow']))
        correlation_j = np.abs(dataframe[feature_j].corr(dataframe['RainTomorrow']))

        # Drop the feature with lower correlation
        if correlation_i < correlation_j:
            dataframe.drop(feature_i, axis=1, inplace=True)
        else:
            dataframe.drop(feature_j, axis=1, inplace=True)

# Drop highly correlated features for numeric features based on target correlation
drop_highly_correlated_features(df_copy, unique_numeric_pairs)

# Now, df_copy contains the original data with one feature dropped from each highly correlated pair
df_copy.head()
```

3.β. Υπάρχουν απουσιάζουσες τιμές; Γράψτε κατάλληλο κώδικα ώστε να χειριστείτε αυτές τις τιμές.

```python
df_copy_new = df_copy.copy()

# Print missing values per column
missing_values = df_copy_new.isnull().sum()
print("Missing values per column before dropping NaN columns:")
print(missing_values)

# Identify features with more than 35% of their values as NaN
features_to_drop = df_copy_new.columns[df_copy_new.isna().mean() > 0.35]

# Identify numeric features with more than 0 and less than 35% of their values as NaN
numeric_features_to_impute_mean = df_copy_new.select_dtypes(include=np.number).columns[
    (df_copy_new.select_dtypes(include=np.number).isna().sum() > 0) &
    (df_copy_new.select_dtypes(include=np.number).isna().mean() < 0.35)
]

# Identify categorical features with more than 0 and less than 35% of their values as NaN
categorical_features_to_impute_mode = df_copy_new.select_dtypes(include=['object']).columns[
    (df_copy_new.select_dtypes(include=['object']).isna().sum() > 0) &
    (df_copy_new.select_dtypes(include=['object']).isna().mean() < 0.35)
]


# Step 2: Drop features with more than 35% of their values as NaN
df_copy_new.drop(columns=features_to_drop, inplace=True)

# Step 3: Impute numeric features with mean
numeric_imputer = SimpleImputer(strategy='mean')
df_copy_new[numeric_features_to_impute_mean] = numeric_imputer.fit_transform(df_copy_new[numeric_features_to_impute_mean])

# Step 4: Impute categorical features with most frequent value
categorical_imputer = SimpleImputer(strategy='most_frequent')
df_copy_new[categorical_features_to_impute_mode] = categorical_imputer.fit_transform(df_copy_new[categorical_features_to_impute_mode])


# Now, check if missing values are handled
missing_values_after_dropping = df_copy_new.isnull().sum()
print("Missing values after dropping NaN columns:")
print(missing_values_after_dropping)

# Print DataFrame information
print(df_copy_new)
```

3.γ. Γράψτε κώδικα για την κατάλληλη μετατροπή των κατηγορικών μεταβλητών ώστε να μπορούν να τους διαχειριστούν οι ταξινομητές που θα χρησιμοποιήσετε.

```python
df_copy_new_1 = df_copy_new.copy()

# Identify categorical features
categorical_features = df_copy_new_1.select_dtypes(include=['object']).columns

# Create a dictionary to store the mapping of labels to original values for each feature
label_mapping = {}

# Apply label encoding to each categorical feature and print the mapping
for feature in categorical_features:
    unique_values = df_copy_new_1[feature].unique()

    # Skip features with only one unique value
    if len(unique_values) > 1:
        label_encoder = LabelEncoder()
        df_copy_new_1[feature] = label_encoder.fit_transform(df_copy_new_1[feature])

        # Store the mapping in the dictionary
        label_mapping[feature] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

# Print the label mappings for each categorical feature
for feature, mapping in label_mapping.items():
    print(f"Label mapping for {feature}:\n{mapping}\n")

df_copy_new_1
```

3.ε. Εκτελέστε όλα τα προηγούμενα βήμα προεκπαίδευσης (χρήση μετασχηματιστών) ώστε να είναι "καθαρό" το αρχικό σύνολο για να χρησιμοποιηθεί για την εκπαίδευση των ταξινομητών.

```python
from sklearn.preprocessing import MinMaxScaler

df_copy_new_2=df_copy_new_1.copy()

# Define the categorical and numeric features
categorical_columns = df_copy_new_2.select_dtypes(include=['object']).columns
numeric_columns = df_copy_new_2.select_dtypes(include=['number']).columns

# Create transformers
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')), #we replace NaN values with the mean
    ('scaler', MinMaxScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')), #we replace NaN values with the most_frequent
    ('label_encoder', LabelEncoder())
])

# Create a column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_columns),
        ('cat', categorical_transformer, categorical_columns)
    ])

# Fit and transform the data
transformed_data = preprocessor.fit_transform(df_copy_new_2)

# Get feature names after transformation
feature_names = np.concatenate([numeric_columns, categorical_columns])

# Create a new DataFrame with the transformed data
df_transformed = pd.DataFrame(transformed_data, columns=feature_names)

print(df_transformed)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import f1_score, make_scorer, classification_report
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt


# Data preprocessing
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_copy)
X_val_scaled = scaler.transform(X_val_copy)

'''
# Model configuration
classifiers = [
('Naive Bayes', GaussianNB(), {'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]}),

('KNN', KNeighborsClassifier(), {'n_neighbors': [3, 5, 7]}),

('Logistic Regression', LogisticRegression(max_iter=3000), {'C': [0.1, 1.0, 10.0], 'penalty': ['l1', 'l2'], 'solver': ['liblinear', 'saga']}),

('MLP', MLPClassifier(max_iter=500), {'hidden_layer_sizes': [(50,), (100,), (150,)], 'activation': ['relu', 'tanh', 'logistic'], 'alpha': [0.0001, 0.001, 0.01]}),

('SVC', SVC(), {'C': [0.1, 1.0, 10.0], 'kernel': ['linear', 'rbf'], 'gamma': ['scale', 'auto']}),

('Decision Tree', DecisionTreeClassifier(), {'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4]}),

('Random Forest', RandomForestClassifier(), {'n_estimators': [50, 100, 150], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4]})

]
'''

# Additional parameters for each algorithm
classifiers = [
    ('Naive Bayes', GaussianNB(), {'var_smoothing': [1e-11, 1e-10, 1e-9, 1e-8, 1e-7, 1e-6, 1e-5]}),

    ('KNN', KNeighborsClassifier(), {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance'], 'p': [1, 2]}),

    ('Logistic Regression', LogisticRegression(max_iter=3000), {'C': [0.1, 1.0, 10.0], 'penalty': ['l1', 'l2'], 'solver': ['liblinear', 'saga'], 'class_weight': [None, 'balanced']}),

    ('MLP', MLPClassifier(max_iter=500), {'hidden_layer_sizes': [(50,), (100,), (150,)]}),

    ('SVC', SVC(), {'C': [0.1, 1.0, 10.0], 'kernel': ['linear', 'rbf']}),

    ('Decision Tree', DecisionTreeClassifier(), {'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4], 'criterion': ['gini', 'entropy'], 'splitter': ['best', 'random']}),

    ('Random Forest', RandomForestClassifier(), {'n_estimators': [50, 100, 150], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4], 'criterion': ['gini', 'entropy']})
]

# Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scorer = make_scorer(f1_score, average='weighted')

# Model training and evaluation
f1_scores = []
best_classifier = None
best_f1_score = 0.0

for name, classifier, param_grid in classifiers:
    # Create GridSearchCV instance
    grid_search = GridSearchCV(classifier, param_grid, scoring=scorer, cv=cv, n_jobs=-1)

    print(name)
    try:
        # Fit the model
        grid_search.fit(X_train_scaled, y_train_copy)

        # Get the best parameters
        best_params = grid_search.best_params_

        # Get the best model
        best_model = grid_search.best_estimator_

        # Evaluate the performance on the validation set
        y_val_pred = best_model.predict(X_val_scaled)
        f1_val = f1_score(y_val_copy, y_val_pred, average='weighted')

        # Print results
        print(f"Best Parameters for {name}: {best_params}")
        print(f"F1 Score on Validation Set for {name}: {f1_val}\n")

        # Store F1 score for later comparison
        f1_scores.append((name, f1_val))

        # Update best classifier if the current model has a higher F1 score
        if f1_val > best_f1_score:
            best_f1_score = f1_val
            best_classifier = best_model

    except Exception as e:
        print(f"An error occurred for {name}: {str(e)}\n")

# Create a bar plot of F1 scores
plt.figure(figsize=(10, 6))
plt.bar([name for name, _ in f1_scores], [f1 for _, f1 in f1_scores], color='skyblue')
plt.title('F1 Scores for Different Models on Validation Set')
plt.xlabel('Model')
plt.ylabel('F1 Score')
plt.ylim(0, 1)  # Set the y-axis limit between 0 and 1 for F1 score
plt.show()

# Return the best classifier
print("Best Classifier:")
print(best_classifier)
```
